```python
#all urls 
base_url='https://www.instagram.com/'
driver.execute_script('window.scrollBy(0,2000);')
time.sleep(5)
all_urls=[]
post=bs(driver.page_source,'lxml')
for i in list(post.find_all(class_='v1Nh3'))[0:30]:
    #print(base_url+i.a['href'])
    all_urls.append(base_url+i.a['href'])
len(all_urls)
```

#### Liking now

```python
driver.refresh()
```

```python
for i in all_urls:
    driver.get(i)
    checklike=bs(driver.page_source,'lxml')
    
    
    time.sleep(2)
    check=checklike.find(class_='_8-yf5')
    if check['aria-label']=='Like':
        print('Photo already liked')
    else:
        print('photo already unliked')
    like=driver.find_element_by_class_name('fr66n')
    like.click()
    
```
